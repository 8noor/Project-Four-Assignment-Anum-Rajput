Simulate rolling two dice, three times. Prints the results of each die roll. This program is used to show how variable scope works.

```python
import random
NUM_SIDES = 6
```

ðŸ’¡ Program Ka Maqsad: Ye program teen dafa do dice roll karta hai (2 dice * 3 baar = 6 rolls). Har dafa total print karta hai. Aur iske zariye hum variable scope (local vs global) ka concept dekhte hain.

random aur NUM_SIDESðŸŽ²

random library ka use hota hai random number generate karne ke liye â€” yahan dice ke rolls ke liye.

NUM_SIDES = 6 ka matlab har die (singular of dice) ke 6 sides hain (jaise normal dice hoti hai).

ðŸŽ² roll_dice() Function

```python
def roll_dice():
    die1: int = random.randint(1, NUM_SIDES)
    die2: int = random.randint(1, NUM_SIDES)
    total: int = die1 + die2
    print("Total of two dice:", total)
```

Ye function 2 dice ko roll karta hai:

die1 aur die2 ko random.randint(1, 6) se values milti hain.

**ðŸš€ main() Function**

```python
def main():
    die1: int = 10
    print("die1 in main() starts as: " + str(die1))
    roll_dice()
    roll_dice()
    roll_dice()
    print("die1 in main() is: " + str(die1))
```

die1 = 10 se ek local variable die1 define hota hai jo main() ke andar hai.

Pehle ye print hota hai: "die1 in main() starts as: 10"

Fir roll_dice() 3 times call hota hai, jisme alag die1 hota hai.

Last line fir se wahi original die1 ko print karti hai.

**ðŸ¤¯ Important Point:**
main() ka die1 = 10 kisi bhi tarah roll_dice() ke die1 ko affect nahi karta.

roll_dice() ke die1 alag hain. Ye scope ka magic hai!

**Notice:**
roll_dice() ke die1 print nahi ho rahe â€” sirf total print ho raha hai.

Lekin main() ka die1 constant hai: 10 hi rahta hai â€” iska matlab roll_dice() ka die1 alag hai.



```python
"""
Program: dicesimulator
----------------------
Simulate rolling two dice, three times.  Prints
the results of each die roll.  This program is used
to show how variable scope works.
"""

# Import the random library which lets us simulate random things like dice!
import random

# Number of sides on each die to roll
NUM_SIDES = 6

def roll_dice():
    """
    Simulates rolling two dice and prints their total
    """
    die1: int = random.randint(1, NUM_SIDES)
    die2: int = random.randint(1, NUM_SIDES)
    total: int = die1 + die2
    print("Total of two dice:", total)

def main():
    die1: int = 10
    print("die1 in main() starts as: " + str(die1))
    roll_dice()
    roll_dice()
    roll_dice()
    print("die1 in main() is: " + str(die1))

# This provided line is required at the end of a Python file
# to call the main() function.
if __name__ == '__main__':
    main()
```

**âœ… Example Program: Student Score Simulator**

```python
# Global variable
bonus_points = 5

def calculate_score():
    # Local variables (sirf is function ke andar kaam kareinge)
    math_score = 80
    science_score = 90
    total = math_score + science_score + bonus_points  # bonus_points global hai
    print("Inside calculate_score():")
    print("Math:", math_score)
    print("Science:", science_score)
    print("Total (with bonus):", total)

def main():
    math_score = 100  # Yeh local hai sirf main() ke liye
    print("Inside main():")
    print("Math score in main():", math_score)

    calculate_score()

    # Dekho yahan math_score ki value wahi 100 hai
    print("Back in main(), math score is still:", math_score)

# Start program
if __name__ == '__main__':
    main()
```
